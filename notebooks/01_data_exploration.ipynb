# Data Exploration\n

This notebook explores the sample datasets for the Intelligent Student Risk Monitoring & Decision Support System.\n
\n
We will load the datasets, examine their structure, check for missing values, and perform basic statistical analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set display options\n
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)

# Load datasets\n
students_df = pd.read_csv('../datasets/students_sample.csv')
marks_df = pd.read_csv('../datasets/marks_sample.csv')
attendance_df = pd.read_csv('../datasets/attendance_sample.csv')
fees_df = pd.read_csv('../datasets/fees_sample.csv')

print('Datasets loaded successfully!')
print(f'Students shape: {students_df.shape}')
print(f'Marks shape: {marks_df.shape}')
print(f'Attendance shape: {attendance_df.shape}')
print(f'Fees shape: {fees_df.shape}')
```
